```python
import json
import numpy as np
import matplotlib
matplotlib.use('TkAgg')
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open('./mount_dir/timing.json', 'r') as fp:
    data = json.load(fp)
```

```python
fig  = plt.figure(figsize=(10,8))
plt.plot(0 * np.ones(10,), (data["fa"]), 'o', label="FA")
plt.plot(1 * np.ones(10,), (data["fa_100"]), 'o', label="FA (sim)")
plt.plot(2 * np.ones(10,), (data["likelihood"]), 'o', label="Likelihood")
plt.legend()
plt.title("Time to evaluate 1000 samples")
plt.xticks([0, 1, 2], ["FA", "FA (sim)", "L"])
plt.ylim([0, 5])
plt.ylabel("Time (s)")
plt.grid()
fig.savefig("./plots/timing.png")
```

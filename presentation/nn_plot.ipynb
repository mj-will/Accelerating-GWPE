```python
from graphviz import Digraph
```

```python
green = "#2ecc71"
red = "#e74c3c"
blue = "#3498db"
black = "#000000"
label = "aaaaaaaa"
```

```python
g = Digraph('g', filename="./images/simple_network.gv");
g.graph_attr.update(splines="false", nodesep='1', ranksep='2');

with g.subgraph(name="clust_layer") as c:
    c.attr(color='white')
    c.attr(rank='same');
    for i in range(0, 3):
        c.node("input" + str(i), color=green, style="filled", fontcolor=green, shape="circle", label=label);
    c.attr(label="Input nodes")
    
with g.subgraph(name="Hidden layer 1") as c:
    c.attr(color='white')
    c.attr(rank='same');
    c.attr(label="Hidden layer",)
    for i in range(0, 4):
        c.node("dense" + str(i), color=blue, style="filled", fontcolor=blue, shape="circle", label=label);
        g.edge("input0", "dense" + str(i))
        g.edge("input1", "dense" + str(i))
        g.edge("input2", "dense" + str(i))
    

with g.subgraph(name="Hidden layer 2") as c:
    c.attr(color='white')
    c.attr(rank='same');

    for i in range(0, 4):
        c.node("dense1" + str(i), color=blue, style="filled", fontcolor=blue, shape="circle", label=label);
        for j in range(0, 4):
            g.edge("dense" + str(j), "dense1" + str(i))
    c.attr(label="Hidden layer", labeljust="right")

with g.subgraph(name="Output layer") as c:
    c.attr(color='white')
    c.attr(rank='same');
    c.node("output", color=red, style="filled", shape="circle", fontcolor=red, label=label);
    for j in range(0, 4):
        g.edge("dense1" + str(j), "output")
    c.attr(label='Output Layer', labelloc="bottom");
g.attr(arrowShape="none");
g.edge_attr.update(arrowhead="none", color="#707070");
```

```python
g.render(view=False)
g
```
